```python
import FinanceDataReader as fdr
from pykrx import stock
# import pandas_datareader as pdr
import pandas_datareader.data as web
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df1 = fdr.DataReader('KS11', '2018-01-01')
df2 = fdr.DataReader('114260', '2018-01-01') # KODEX 국고채3년

# 컬럼명 정리
df1 = df1[['Close']]
df1.rename(columns={'Close': 'market_kospi'}, inplace=True)
df2 = df2[['Close']]
df2.rename(columns={'Close': 'bond_k3y'}, inplace=True)

df = df1.merge(df2, left_index=True, right_index=True, how='inner')
df = df.rename_axis('date')

df.head()
```

```python
# 종가 이동평균으로 나눈 현재 종가 비율
df['market_kospi_ma5_ratio'] = df['market_kospi'] / df['market_kospi'].rolling(window=5).mean()
df['market_kospi_ma20_ratio'] = df['market_kospi'] / df['market_kospi'].rolling(window=10).mean()
df['market_kospi_ma60_ratio'] = df['market_kospi'] / df['market_kospi'].rolling(window=60).mean()
df['market_kospi_ma120_ratio'] = df['market_kospi'] / df['market_kospi'].rolling(window=120).mean()

# 거래량 이동평균으로 나눈 현재 거래량 비율
df['bond_k3y_ma5_ratio'] = df['bond_k3y'] / df['bond_k3y'].rolling(window=5).mean()
df['bond_k3y_ma20_ratio'] = df['bond_k3y'] / df['bond_k3y'].rolling(window=10).mean()
df['bond_k3y_ma60_ratio'] = df['bond_k3y'] / df['bond_k3y'].rolling(window=60).mean()
df['bond_k3y_ma120_ratio'] = df['bond_k3y'] / df['bond_k3y'].rolling(window=120).mean()
df.head()
```

```python
# 결측값을 포함하는 행 제거
df = df.dropna()
df.head()
```
